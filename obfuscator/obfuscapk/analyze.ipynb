```python
import pandas as pd

df_f = pd.read_csv('./stats/FieldRename.csv', index_col='hash')
df_m = pd.read_csv('./stats/MethodRename.csv', index_col='hash')
df_c = pd.read_csv('./stats/ClassRename.csv', index_col='hash')

df_f.head(5)
```

```python
f_true = df_f[df_f.obfuscated == True]
f_false = df_f[df_f.obfuscated == False]
m_true = df_m[df_m.obfuscated == True]
m_false = df_m[df_m.obfuscated == False]
c_true = df_c[df_c.obfuscated == True]
c_false = df_c[df_c.obfuscated == False]

print('\nFieldRename')
print('### True =', len(f_true))
print('### False =', len(f_false))

print('\nMethodRename')
print('### True =', len(m_true))
print('### False =', len(m_false))

print('\nClassRename')
print('### True =', len(c_true))
print('### False =', len(c_false))
```

```python
import matplotlib.pyplot as plt

def plot_boxplot(time_data, title):
    plt.figure(figsize=(10, 4))
    plt.boxplot(time_data, labels=['Field Renaming', 'Method Renaming', 'Class Renaming'])
    plt.title(f'Obfuscapk - {title}')
    plt.ylabel('Time (seconds)')
    plt.grid(True)
    plt.show()
```

```python
time_data_true = [f_true['time'], m_true['time'], c_true['time']]
time_data_false = [f_false['time'], m_false['time'], c_false['time']]

plot_boxplot(time_data_true, 'Time for successful obfuscation')
plot_boxplot(time_data_false, 'Time for failed obfuscation')
```
